```python
url = "https://api.scb.se/OV0104/v1/doris/en/ssd/START/AM/AM0401/AM0401L/NAKUArblheltidstudAr"

params = {
    "query": [
        {
            "code": "Arbetskraftstillh",
            "selection": {
                "filter": "item",
                "values": ["ALÖS"]
            }
        },
        {
            "code": "Kon",
            "selection": {
                "filter": "item",
                "values": ["1+2"]
            }
        },
        {
            "code": "Alder",
            "selection": {
                "filter": "item",
                "values": ["tot15-74"]
            }
        },
        {
            "code": "ContentsCode",
            "selection": {
                "filter": "item",
                "values": ["AM04011Q"]
            }
        },
        {
            "code": "Tid",
            "selection": {
                "filter": "item",
                "values": ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021","2022","2023","2024","2025"]
            }
        }
    ],
    "response": {
        "format": "json"
    }
}

response_data = requests.post(url, json=params)
print("Status:", response_data.status_code)
print(response_data.json())
```

```python
data = response_data.json()

years = []
unemployment_rate = []

for item in data['data']:
    years.append(int(item['key'][-1]))
    unemployment_rate.append(float(item['values'][0]))

df_unemployment = pd.DataFrame({
    'year': years,
    'unemployment_rate': unemployment_rate
})

print(df_unemployment)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df_unemployment['year'], df_unemployment['unemployment_rate'],
         marker='o',
         linewidth=2.5,
         color='steelblue',
         markersize=8)

plt.axvspan(2020, 2021, color='red', alpha=0.1, label='COVID-19')
plt.axvline(x=2022, color='green', linestyle='--', linewidth=1.5);
plt.text(2022.1, 9.0, 'Post-COVID\nRecovery', color='green', fontsize=10)

plt.title('Sweden Unemployment Rate (2010-2025)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.grid(True, alpha=0.4)
plt.legend()
plt.tight_layout()
plt.savefig('sweden_unemployment.png', dpi=150)
plt.show()
```

```python
url = "https://api.scb.se/OV0104/v1/doris/en/ssd/START/AM/AM0401/AM0401L/NAKUArblheltidstudAr"

params_youth = {
    "query": [
        {"code": "Arbetskraftstillh", "selection": {"filter": "item", "values": ["ALÖS"]}},
        {"code": "Kon", "selection": {"filter": "item", "values": ["1+2"]}},
        {"code": "Alder", "selection": {"filter": "item", "values": ["15-24"]}},
        {"code": "ContentsCode", "selection": {"filter": "item", "values": ["AM04011Q"]}},
        {"code": "Tid", "selection": {"filter": "item", "values": ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021","2022","2023","2024","2025"]}}
    ],
    "response": {"format": "json"}
}

response_youth = requests.post(url, json=params_youth)

years_youth = []
unemployment_youth = []

for item in response_youth.json()['data']:
    years_youth.append(int(item['key'][-1]))
    unemployment_youth.append(float(item['values'][0]))

df_youth = pd.DataFrame({
    'year': years_youth,
    'unemployment_rate': unemployment_youth
})

print(df_youth)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df_unemployment['year'], df_unemployment['unemployment_rate'],
         marker='o', linewidth=2.5, color='steelblue', 
         markersize=8, label='General (15-74)')

plt.plot(df_youth['year'], df_youth['unemployment_rate'],
         marker='o', linewidth=2.5, color='coral',
         markersize=8, label='Youth (15-24)')

plt.axvspan(2020, 2021, color='red', alpha=0.1, label='COVID-19')

plt.title('Sweden Unemployment Rate: General vs Youth (2010-2025)', 
          fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.4)
plt.tight_layout()
plt.savefig('unemployment_comparison.png', dpi=150)
plt.show()
```

```python
print("=" * 45)
print("📊 SWEDEN UNEMPLOYMENT SUMMARY")
print("=" * 45)

print(f"\nGeneral Unemployment (15-74):")
print(f"  Average : {df_unemployment['unemployment_rate'].mean():.1f}%")
print(f"  Lowest  : {df_unemployment['unemployment_rate'].min():.1f}% ({df_unemployment.loc[df_unemployment['unemployment_rate'].idxmin(), 'year']})")
print(f"  Highest : {df_unemployment['unemployment_rate'].max():.1f}% ({df_unemployment.loc[df_unemployment['unemployment_rate'].idxmax(), 'year']})")

print(f"\nYouth Unemployment (15-24):")
print(f"  Average : {df_youth['unemployment_rate'].mean():.1f}%")
print(f"  Lowest  : {df_youth['unemployment_rate'].min():.1f}% ({df_youth.loc[df_youth['unemployment_rate'].idxmin(), 'year']})")
print(f"  Highest : {df_youth['unemployment_rate'].max():.1f}% ({df_youth.loc[df_youth['unemployment_rate'].idxmax(), 'year']})")

print(f"\nYouth unemployment is on average")
print(f"{df_youth['unemployment_rate'].mean() / df_unemployment['unemployment_rate'].mean():.1f}x higher than general!")
```
